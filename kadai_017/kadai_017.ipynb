```python
class Human:
    def __init__(self,name,age):
      self.name = name
      self.age = age
    def check_adult(self):
      if self.age >= 18:
        print(f"{self.name}は大人です")
      else:
        print(f"{self.name}は未成年です")

member = {"太郎":1, "次郎":2,"三郎":30}

for name,age in member.items():
  human = Human(name,age)
  human.check_adult()
```
