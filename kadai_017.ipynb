```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です ")
    else:
      print(f"{self.name}は大人ではない")

taro = Human("太郎",20)
jiro = Human("次郎",15)
saburo = Human("三郎",5)

member = [taro,jiro,saburo]

for person in member:
  person.check_adult()
```
